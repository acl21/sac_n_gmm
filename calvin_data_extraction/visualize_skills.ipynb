```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
```

```python
demos_path = '/home/lagandua/calvin_demos/'
skills = glob.glob(os.path.join(demos_path, '*/*.npy'))
```

```python
skills = {x.split('/')[-2]: os.path.join(demos_path, f'{x.split("/")[-2]}/training.npy') for x in skills}
```

```python
def plot_3d_trajectories(demos, repro=None):
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(projection="3d")
    if repro is None:
        for i in range(demos.shape[0]):
            x_val = demos[i, :, 0]
            y_val = demos[i, :, 1]
            z_val = demos[i, :, 2]
            ax.scatter(x_val, y_val, z_val, s=10)
    else:
        ax.scatter(demos[:, 0], demos[:, 1], demos[:, 2], alpha=0.5, s=1, label="Demonstration")
        ax.scatter(repro[:, 0], repro[:, 1], repro[:, 2], s=5, label="Reproduction")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
skill_data = np.load(skills['move_slider_left'])
plot_3d_trajectories(skill_data)
```
